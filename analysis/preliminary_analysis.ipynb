```python
from pathlib import Path
import numpy as np
import h5py

fname = Path("data/dset.h5")

with h5py.File(fname, 'r') as f:
    #f = h5py.File(fname, 'r')
    list(f.keys())
    dset = f["data"]
    print(dset.shape)

    t_1, t_2 = int(0*dset.shape[0]) , dset.shape[0] 
    t_step = 1
    dset_arr = dset[t_1:t_2:t_step, :, 0]
```

```python
L = 230
t = int((t_2 - t_1)/t_step) - 1
lat = np.zeros(L, dtype=int)
for x in dset_arr[t, :]:
    lat[x] += 1

lat
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(1, 1, figsize=(14, 4))
ax.bar(range(len(lat)), lat)
ax.set_xlim(-0.5, len(lat)-0.5)
ax.set_title(r"hop rate $bn^{-1.1}$")
plt.savefig("example_1overnto1point1.png", format="png")
```
